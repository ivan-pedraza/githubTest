```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from datetime import date
import datetime, time
from time import gmtime, strftime
from sklearn.preprocessing import StandardScaler
#import xgboost as xgb
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
import pyreadr
import math
import numpy as np
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df_sales = pyreadr.read_r('Data/venta_interna_4.rds')
sales = df_sales[None]
encuestas = pd.read_csv('Data/Base Datos Tienda Perfecta Proximity.csv', low_memory=False)
fechas = pd.read_csv('Data/fechas.csv')
```

```python
fechas_dia = fechas[['id_dia','dia_fecha']]
sales['id_dia'] = sales['id_dia'].astype(int)
sales_2 = sales.merge(fechas_dia,on='id_dia',how='inner')
sales_3 = sales_2[['id_cli_corp','dia_fecha','des_unidnegocio','pesos','unidades']]
```

```python
sales_4 = sales_3.loc[sales_3['pesos'] > 0]
sales_4 = sales_4.reset_index()
sales_4 = sales_4.drop('index',axis=1)
```

```python
encuestas['Start time'] = pd.to_datetime(encuestas['Start time'],format='%d/%m/%Y')
sales_4['dia_fecha'] = pd.to_datetime(sales_4['dia_fecha'],format='%d/%m/%Y')
```

```python
encuestas_2 = encuestas[encuestas.columns[~encuestas.columns.str.contains('Razón')]]
encuestas_2 = encuestas_2[['Digita el número de cliente SAP Pepsico','Start time','Total points','Porcentaje',
                        'Espacio en mueble salado', 'Rotación Salado', 'Precios Salado',
                        'Espacio en mueble galleta',
                        'Rotación Galleta', 'Precios Galleta', 'Parrilla Quaker',
                        'Ganchos cacahuate', 'Frentes en góndola y checkout',
                        'Exhibición primaria salado', 'Exhibición secundaria salado',
                        'Comida rápida', 'Exhibición Nueces & Semillas',
                        'Exhibición primaria galleta', 'Exhibición secundaria galleta',
                        'Área de café', 'Salado ó galleta - Checkout', 'Batalla regional','Comunicación']]
#'Porcentaje góndola salado''Porcentaje góndola galleta'
```

```python
encuestas_2 = encuestas_2.fillna(0)
```

```python
encuestas_2 = encuestas_2.rename(columns={'Digita el número de cliente SAP Pepsico':'id_tienda'})
encuestas_2 = encuestas_2.rename(columns={'Start time':'dia_fecha'})
encuestas_2['id_tienda'] = encuestas_2['id_tienda'].str.slice(0,-3)
#Se consideran solo aquellos registros donde el id de la tienda sea de longitud de 9
encuestas_2['len_id'] = encuestas_2['id_tienda'].str.len()
encuestas_3 = encuestas_2.loc[encuestas_2['len_id'] == 9]
#Se resetea indice para evitar SettingWithCopyWarning
encuestas_3 = encuestas_3.reset_index()
encuestas_3 = encuestas_3.drop('index',axis=1)
# Se agrega al inicio un 0 para cada ID
encuestas_3['id_tienda'] = '0' + encuestas_3['id_tienda']
```

```python
sales_g = sales_4.groupby(['id_cli_corp','dia_fecha','des_unidnegocio']).agg(Sum_Venta=('pesos','sum'),
                                                                             Sum_Unidades=('unidades','sum')).reset_index()
```

```python
pivot_sales = sales_g.pivot_table(values='Sum_Venta',index=['id_cli_corp','dia_fecha'],columns='des_unidnegocio',aggfunc=np.sum).reset_index().rename_axis(None, axis=1)
pivot_sales = pivot_sales.fillna(0)
pivot_sales = pivot_sales.rename(columns={'id_cli_corp':'id_tienda'})
```

```python
info = encuestas_3.merge(pivot_sales,on=['id_tienda','dia_fecha'],how='inner')
info = info.sort_values(by=['id_tienda','dia_fecha'])
info = info.reset_index()
info = info.drop(['index','len_id'],axis=1)
```

```python
info['Venta Total'] = info['BISCUITS'] + info['CONFECTIONERY & NB'] + info['FOODS'] + info['SAVORY SNACKS']
```
